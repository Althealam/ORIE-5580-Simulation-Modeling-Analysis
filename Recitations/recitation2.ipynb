##Some standard plotting  functions

###Using axis, title, and colour

Functions:  plt.xlabel(), plt.ylabel(), plt.title()

Easiest way to set colour is to set it to one of these codes when using plt.plot():  {'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'}.
More complicated ways can be found here: 
https://matplotlib.org/gallery/color/color_demo.html#sphx-glr-gallery-color-color-demo-py


```python
#Suppose these are our x, y vectors to be plotted
x = np.arange(0, 4*math.pi, 0.01)
y = np.sin(x)

#Plotting with a red line.
#Note that we could have left out color = 'r' and plt would just choose the colour for us
plt.plot(x,y, color = 'r')

#adding labels and titles
plt.xlabel('x is here')
plt.ylabel('y is here')
plt.title('My RED sin Curve')
```

###Plotting two lines on the same plot, with a legend

```python
#Here, y2 = sin(2x).  We want to plot sin(x) and sin(2x).  These can be labeled
#with a legend.
y2 = np.sin(2*x) 
plt.plot(x, y, label='sin(x)')
plt.plot(x, y2, label='sin(2x)') #we have to tell plt to put a legend on!
plt.legend(loc=1)
plt.xlabel('x is here')
plt.ylabel('y is here')

plt.title('My DOUBLE sin Curves')
plt.show()
```

###Scatterplots:  Changing the type of points

You can find more about scatterplots here:
https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html



```python
#We reduce the number of values at which we compute sin(x) to make the points more obvious
x = np.arange(0, 4*math.pi, 0.1)
y = np.sin(x)

#Basic markers (i.e. type of points): ('o', 'v', '^', '<', '>', '8', 's', 'p', '*', 'h', 'H', 'D', 'd', 'P', 'X') 
#s sets the size
plt.scatter(x,y, marker = 's', s=10, color = 'm')

#adding labels and titles
plt.xlabel('x is here')
plt.ylabel('y is here')
plt.title('Scatterplot Version of sin Curve')
```
