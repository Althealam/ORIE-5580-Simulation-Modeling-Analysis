**c.** Define $N_{circle}$ to be the (random) number of points which lie within the circle. What is $\mathbb{E}[N_{circle}]$ and $Var(N_{circle})$ (express your answer in terms of $\pi$)?

Suppose also that you did not know the value of $\pi$; can you still argue that $Var(N_{circle})\leq \frac{n}{4}$?


---

### Ans.


Let $N_i$ be a Bernoulli random variable with probability of success $\frac{\pi}{4}$. Notice that we have:
$$N_{circle} = \sum_{i = 1}^{n} N_i$$
Notice that for bernouli random variable, we must have $\mathbb{E}[N_i] = \frac{\pi}{4}$ and $Var[N_i] = \frac{\pi}{4}(1-\frac{\pi}{4})$. We can use linearity of expectation to compute:
$$\mathbb{E}[N_{circle}] = \mathbb{E}[\sum_{i = 1}^{n} N_i] = \sum_{i = 1}^{n} \mathbb{E}[ N_i] = n\cdot \frac{\pi}{4}$$
Since $N_i$ is drawn independently, we also have:
$$Var[N_{circle}] = Var[\sum_{i = 1}^{n} N_i] = \sum_{i = 1}^{n} Var[N_i] = n \cdot \frac{\pi}{4}(1-\frac{\pi}{4})$$.
Lastly, notice that a function $f(x) = x(1-x)$ is a concave function with maximum value to be $1/4$ at $x = 1/2$, then we must have $\frac{\pi}{4}(1-\frac{\pi}{4}) \leq \frac{1}{4}$, so $Var[N_{circle}]\leq \frac{n}{4}$.

**d.** Given $n$ random points generated as in part (b), we can estimate $\pi$ as $\pi_n = \frac{4N_{circle}}{\pi}$ for $\pi$ (where as before $N_{circle}$ is the random number of points that lie inside the circle). For $n$ between $1$ and $1000$, plot the relative error $|\pi_n - \pi|/\pi$. Also plot the function $1/\sqrt{n}$ on the same plot. How do the relative errors compare to this function?

### Ans.


```python
#number of samples
N_array = np.arange(9,1000,1)+1
box_area = 4.0    
pi_estimate = np.array([])

#initializing seed for random number generator
#np.random.seed(110)

for N_total in N_array:
    X = np.random.uniform(low=-1, high=1, size=N_total)  
    Y = np.random.uniform(low=-1, high=1, size=N_total)   
    # sum up the hits inside the circle
    N_inside=np.sum(np.sqrt(X**2+Y**2)<1.0)
    # estimate the circle area
    pi_estimate = np.append(pi_estimate, box_area * N_inside/N_total)

estimate_errors = np.abs(pi_estimate - np.pi)/np.pi    
    
# visualizing the relative estimation error
plt.plot(N_array,estimate_errors,'.')  
plt.plot(N_array,1.0/np.sqrt(N_array),':r')  

plt.xlabel('number of samples')
plt.ylabel('relative estimation error')
plt.title('Relative estimation errors vs. number of samples')
plt.show()
```

Above we used simulation to compute the area of a unit circle. Using a similar process, we can find the volume of a 'sphere' in $n$ dimensions. Estimate this volume for dimensions $d\in\{2,3,\ldots,20\}$, using $100,000$ samples in each dimension. How does your estimate of the volume of the sphere behave as the dimension increases? Also, do your estimates seem smooth or noisy (and if the latter, why)?

### Ans.

```python
#number of samples
N_total = 100000

dim_array = np.arange(1,21)

#initializing seed for random number generator
#np.random.seed(110)
    
volume_estimate = np.array([])
true_volume = np.array([])
for dim in dim_array:
    box_volume = 2**dim
    X = np.random.uniform(low=-1, high=1, size=(dim,N_total))  
    N_distance = np.sqrt(np.sum(X**2,0))
    N_inside = np.sum(N_distance<1.0)
    volume_estimate = np.append(volume_estimate, box_volume * N_inside/N_total)

    # Computing the volume of a hypersphere from formula
    # Refer wikipedia: https://en.wikipedia.org/wiki/Volume_of_an_n-ball
    true_volume = np.append(true_volume,(np.pi)**(dim/2)/sc.special.gamma(1+dim/2))
    
plt.plot(dim_array,volume_estimate,'.')  
plt.plot(dim_array,true_volume,'g:')  
plt.title('Estimating volume of hypersphere in higher dimensions')
plt.xlabel('dimension')
plt.ylabel('volume of hypersphere')
plt.show()
```
