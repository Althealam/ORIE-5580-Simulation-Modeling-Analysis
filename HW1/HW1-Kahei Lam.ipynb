#### Final Result  
Combining all cases, the CDF is:  

$$
F_X(x) =

\begin{align*}
X = \begin{cases} 1 & \text{with probability $p$} \\ 0 & \text{with probability $1-p$} \end{cases}

---


**a.** What is $Cov( X, Y )$?


---


### Ans.
The covariance of $ X $ and $ Y $ is defined as:  
$$
\text{Cov}(X, Y) = E[XY] - E[X]E[Y]
$$

For Bernoulli random variables, the expectation equals the success probability:  
$$
E[X] = P(X = 1) = p, \quad E[Y] = P(Y = 1) = q
$$  

The product $ XY $  is 1 only if both $ X = 1 $ and $ Y = 1 $; otherwise, it is 0:  
$$
E[XY] = P(XY = 1) = P(X = 1, Y = 1) = \frac{1}{3}
$$

Substitute into the covariance formula:  
$$
\text{Cov}(X, Y) = \frac{1}{3} - pq
$$

---


### Ans.

Random variables are uncorrelated if their covariance is 0:  
$$
\text{Cov}(X, Y) = 0 \implies \frac{1}{3} - pq = 0
$$

Thus:  
$$
pq = \frac{1}{3}
$$

## Q3:
In Chapter $1$ example $1.4$, Grinstead and Snell talk about measuring two natural quantities associated with the so-called *simple random walk* $S_0 = 0, S_t = S_{t-1}+ X_t$, where $X_t = \pm 1$ with probability $\frac{1}{2}$ (they describe this as a simple betting game between Peter and Paul, where at each time $t$ they toss a fair coin, and if it comes up heads, then Peter wins a penny from Paul, else Paul wins a penny from Peter). We will now replicate their experiments.

**a.** Generate $n = 1000$ copies of a simple random walk $S^i = (S^i_0,S^i_1,\ldots,S^i_{T})$ (with $T=100$ and $i\in\{1,2,\ldots,n\}$). For each walk, compute $Z^i_T = \frac{S^i_T}{T}$ (i.e., the per-round winnings of Paul after $100$ games), as well as $W^i_T = \frac{1}{T}\sum_{t=1}^{T}\mathbb{1}_{\{S^i_t>0\}}$ (i.e., the fraction of rounds in $[1,T]$ where Paul is in the lead). Plot a histogram of both these quantities (using `plt.hist(data, density=True,bins='auto')`).


---

### Ans.


```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(111)

n = 1000
T = 100

steps = np.random.choice([-1, 1], size=(n, T))
S = np.cumsum(steps, axis=1)
S = np.hstack([np.zeros((n, 1)), S])

Z_T = S[:, -1] / T

W_T = np.mean(S[:, 1:] > 0, axis=1)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(Z_T, bins='auto', density=True)
plt.title(r'Histogram of $Z_T^i = S_T^i / T$')
plt.xlabel(r'$Z_T^i$')
plt.ylabel('Density')
plt.subplot(1, 2, 2)
plt.hist(W_T, bins='auto', density=True)
plt.title(r'Histogram of $W_T^i = \frac{1}{T} \sum_{t=1}^T 1_{\{S_t^i > 0\}}$')
plt.xlabel(r'$W_T^i$')
plt.ylabel('Density')
plt.show()
```

**c.** How do you think the distributions for $Z_T$ and $W_T$ behave as $T→\infty$? Can you give some intuition why this is the case?

### Ans.

As $T \to \infty$, we have:$ Z_T^i \to 0 $ in probability.

Since each $X_t$ has mean $0$ and variance $1$, we have:
$$
\mathbb{E}[S_T] = T \cdot \mathbb{E}[X_t] = 0,
$$
$$
\mathrm{Var}(S_T) = T \cdot \mathrm{Var}(X_t) = T.
$$

So, for large $T$, we can approximate:
$$
S_T \sim N(0, T).
$$

so $Z_T = \frac{S_T^i}{T} \sim N\left(0, \frac{1}{T}\right)$,
which concentrates around 0 as $T$ increases.

As $T \to \infty$, the distribution of $W_T^i$ converges to the arcsine distribution, showned by a U-shaped curve:
$
W_T^i \xrightarrow{d} \text{Arcsine}(0, 1).
$
This means most walks spend almost all or almost none of the time above zero and switching leads is rare. This is because a random walk has momentum — once it drifts above 0, it tends to stay positive for a while, and vice versa.

**c.** Define $N_{circle}$ to be the (random) number of points which lie within the circle. What is $\mathbb{E}[N_{circle}]$ and $Var(N_{circle})$ (express your answer in terms of $\pi$)?

Suppose also that you did not know the value of $\pi$; can you still argue that $Var(N_{circle})\leq \frac{n}{4}$?


---

### Ans.

The area of the square is $4$, and the area of the unit circle is $\pi$, so the probability that a randomly chosen point falls inside the circle is:

$$
p = \frac{\pi}{4}
$$

Since each point falls inside the circle independently with probability $p$, the number of points inside the circle follows a binomial distribution:
$$
N_{\text{circle}} \sim \text{Binomial}(n, p)
$$

The expected number of points inside the circle is:
$$
\mathbb{E}[N_{\text{circle}}] = n \cdot p = n \cdot \frac{\pi}{4}
$$


The variance of the number of points inside the circle is:
$$
\mathrm{Var}(N_{\text{circle}}) = n \cdot p \cdot (1 - p) = n \cdot \frac{\pi}{4} \left(1 - \frac{\pi}{4} \right)
$$


The function $p(1 - p)$ achieves its maximum at $p = \frac{1}{2}$, where $p(1 - p) = \frac{1}{4}$. Therefore, even if we do not know the value of $\pi$, we can bound the variance as:
$$
\mathrm{Var}(N_{\text{circle}}) \leq n \cdot \frac{1}{4}
$$

**d.** Given $n$ random points generated as in part (b), we can estimate $\pi$ as $\pi_n = \frac{4N_{circle}}{n}$ for $\pi$ (where as before $N_{circle}$ is the random number of points that lie inside the circle). For $n$ between $1$ and $1000$, plot the relative error $|\pi_n - \pi|/\pi$. Also plot the function $1/\sqrt{n}$ on the same plot. How do the relative errors compare to this function?


---

### Ans.

The relative error stays below to $1/\sqrt{n}$ from $n = 1$ to $n = 1000$, which is the bound on the error of Monte Carlo methods.

```python
import numpy as np
import matplotlib.pyplot as plt

pi_true = np.pi
n_vals = np.arange(1, 1001)
pi_estimates = np.zeros_like(n_vals, dtype=float)
rel_errors = np.zeros_like(n_vals, dtype=float)

np.random.seed(111)
X = np.random.uniform(-1, 1, size=1000)
Y = np.random.uniform(-1, 1, size=1000)
inside_circle = (X**2 + Y**2 <= 1).astype(int)
cumulative_inside = np.cumsum(inside_circle)

for i, n in enumerate(n_vals):
    pi_n = 4 * cumulative_inside[i] / n
    pi_estimates[i] = pi_n
    rel_errors[i] = np.abs(pi_n - pi_true) / pi_true

plt.figure(figsize=(10, 6))
plt.plot(n_vals, rel_errors, label='Relative Error $|\\pi_n - \\pi| / \\pi$', color='blue')
plt.plot(n_vals, 1 / np.sqrt(n_vals), label='$1/\\sqrt{n}$', color='red')
plt.xlabel("n (number of samples)")
plt.ylabel("Error")
plt.legend()
plt.grid(True)
plt.show()
```

e. Recall that a $95\%$ confidence interval for an estimate is given by $\pi_n \pm \frac{1.96\sqrt{Var(\pi_n)}}{\sqrt{n}}$ (the relative error plot in previous question shows why this is the case). Using your upper bound for $Var(\pi_n)$ from part (c), compute what value of $n$ is required to ensure that the width of the confidence interval is less than $0.001$, and then, simulate using this value $n$ and report an estimate with a $95\%$ confidence interval for $\pi$ that is accurate up to the second digit after the decimal point.




---




### Ans.

```python
import numpy as np
import matplotlib.pyplot as plt

max_width = 0.001
z = 1.96
numerator = 2 * z * 2
required_n = int(np.ceil(numerator / max_width))
print(f"Required sample size n to get CI width < {max_width}: {required_n}")

np.random.seed(111)
X = np.random.uniform(-1, 1, size=required_n)
Y = np.random.uniform(-1, 1, size=required_n)
inside_circle = (X**2 + Y**2 <= 1)
N_circle = np.sum(inside_circle)

pi_hat = 4 * N_circle / required_n

std_error_bound = np.sqrt(4 / required_n)
ci_lower = pi_hat - z * std_error_bound
ci_upper = pi_hat + z * std_error_bound

print(f"\nEstimated π: {pi_hat:.2f}")
print(f"95% Confidence Interval: [{ci_lower:.2f}, {ci_upper:.2f}]")
print(f"Confidence Interval Width: {ci_upper - ci_lower:.2f}")
```

Above we used simulation to compute the area of a unit circle. Using a similar process, we can find the volume of a 'sphere' in $n$ dimensions. Estimate this volume for dimensions $d\in\{2,3,\ldots,20\}$, using $100,000$ samples in each dimension. How does your estimate of the volume of the sphere behave as the dimension increases? Also, do your estimates seem smooth or noisy (and if the latter, why)?

### Ans.

As $d$ increases, the volume shrinks rapidly.
The estimates become noisier in higher dimensions due to fewer points landing inside the ball.
This might because most of the cube's volume lies far from the origin in high dimensions.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(111)

num_samples = 100000
dimensions = range(2, 21)
volumes = []

for d in dimensions:
    X = np.random.uniform(-1, 1, size=(num_samples, d))
    squared_norms = np.sum(X**2, axis=1)
    inside_ball = np.sum(squared_norms <= 1)

    cube_volume = 2 ** d
    ball_volume = (inside_ball / num_samples) * cube_volume
    volumes.append(ball_volume)

plt.figure(figsize=(10, 6))
plt.plot(dimensions, volumes)
plt.title("Estimated Volume of the Unit Ball in $\mathbb{R}^d$")
plt.xlabel("Dimension $d$")
plt.ylabel("Estimated Volume")
plt.grid(True)
plt.show()
```
